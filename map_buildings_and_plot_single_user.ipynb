```python
def plot_single_user(locData):
    locations = locData[['lat', 'long']]
    locationlist = locations.values.tolist()
    map = folium.Map(location=[41.698, -86.24], zoom_start=12)
    for point in range(0, len(locationlist)):
        folium.Marker(locationlist[point], popup=locData['Location_type'][point]).add_to(map)
    return map
```

```python
def map_buildings_to_locations(locData):
    location = []
    loc_type = []
    buildings = pd.read_csv("nd_building_mapping.csv")
    for row in locData.index:
        for ind in buildings.index:
            if (locData.lat[row] <= buildings.max_lat[ind] and locData.lat[row] >= buildings.min_lat[ind]
            and locData.long[row] <= buildings.max_lon[ind] and locData.long[row] >= buildings.min_lon[ind]):
                location.append(buildings.name[ind])
                loc_type.append(buildings.type[ind])
        if(len(location) <= row):
            location.append("None")
            loc_type.append("None")
    locData['Location_name'] = location
    locData['Location_type'] = loc_type
```

```python
map_buildings_to_locations(user1)
user1
```
